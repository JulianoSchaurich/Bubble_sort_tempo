# Estrutura de Dados - Gráfico do Tempo de Ordenaçao por Bubble Sort

Implemente em Python a ordenação pelo método da bolha (Bubble sort) e aplique o método aos seguintes vetores gerados computacionalmente:

*   crescente[100], crescente[1000], crescente[10000], crescente[100000]
*   descrescente[100], decrescente[1000], decrescente[10000], decrescente[100000]
*   aleatório[100], aleatório[1000], aleatório[10000], aleatório[100000]

Para cada um dos 3 casos, plote um gráfico de linha, ou se conseguir pode plotar tudo em uma gráfico só com diferentes cores de linha para cada caso.

```python
from random import sample
import time
import matplotlib.pyplot as plt

def bubbleSort(lista):
    for num in range(len(lista)-1,0,-1):
        for i in range(num):
            if lista[i] > lista[i+1]:
                lista[i], lista[i+1] = lista[i+1], lista[i]
    return lista

def graficoTempoOrdenarvetor(vetores):
  tempo = list(range(len(vetores)))
  tamanhoVetores = list(range(len(vetores)))

  cont = 0
  while cont < len(tamanhoVetores):
    tamanhoVetores[cont] = len(vetores[cont])
    cont+=1
  cont=0

  for i in vetores:
    inicio = time.time()
    bubbleSort(i)
    fim = time.time()
    tempo[cont] = fim - inicio
    cont+=1

  plt.plot(tamanhoVetores, tempo)
  plt.ylabel('Tempo para Ordenar - Bubble')
  plt.xlabel('Tamanho do Vetor')
  plt.show()

#VETORES ORDENADOS
graficoTempoOrdenarvetor([list(range(100)), list(range(1000)), list(range(10000)), list(range(100000))])
#VETORES INVERTIDOS
graficoTempoOrdenarvetor([list(reversed(list(range(100)))), list(reversed(list(range(1000)))), list(reversed(list(range(10000)))), list(reversed(list(range(100000))))])
#VETORES ALEATORIOS
graficoTempoOrdenarvetor([sample(range(1, 101), 100), sample(range(1, 1001), 1000), sample(range(1, 10001), 10000), sample(range(1, 100001), 100000)])
```

```python
from random import sample
import time
import matplotlib.pyplot as plt

def insertionSort(lista):
  for posicao in range(1, len(lista)): 
      key = lista[posicao] 
      # Mover todos os elementos da lista [0..posicao-1] que são maiores que o elemento da vez para as posições à frente dele
      j = posicao-1
      while j >= 0 and key < lista[j] : 
              lista[j+1] = lista[j] 
              j -= 1
      lista[j+1] = key

def selectionSort(lista):
  for posicao in range(len(lista)): 
      # Encontrar o menor elemento da lista 
      posicao_menor = posicao 
      for j in range(posicao+1, len(lista)): 
          if lista[j] < lista[posicao_menor]: 
              posicao_menor = j 
      # Trocar o menor elemento da lista com a posição mais à esquerda ainda não ordenada         
      lista[posicao], lista[posicao_menor] = lista[posicao_menor], lista[posicao] 

def bubbleSort(lista):
    for num in range(len(lista)-1,0,-1):
        for i in range(num):
            if lista[i] > lista[i+1]:
                lista[i], lista[i+1] = lista[i+1], lista[i]
  
def retornarTempoExecucao(lista, tipoOrganizacao):
  if tipoOrganizacao == 'insertion':
    inicio = time.time()
    insertionSort(lista)
    fim = time.time()
  elif tipoOrganizacao == 'selection':
    inicio = time.time()
    selectionSort(lista)
    fim = time.time()
  elif tipoOrganizacao == 'bubble':
    inicio = time.time()
    bubbleSort(lista)
    fim = time.time()
  
  return fim - inicio
      
def retornarX(vetoresOrdenados, vetoresInvertidos, vetoresAleatorios):
  tamanhoVetoresOrdenados, tamanhoVetoresInvertidos, tamanhoVetoresAleatorios = list(range(len(vetoresOrdenados))), list(range(len(vetoresInvertidos))), list(range(len(vetoresAleatorios)))

  tamanho = len(vetoresOrdenados)
  if tamanho > len(vetoresInvertidos):
    tamanho = len(vetoresInvertidos)
  if tamanho > len(vetoresAleatorios):
    tamanho = len(vetoresAleatorios)

  cont = 0
  while cont < tamanho:
    tamanhoVetoresOrdenados[cont], tamanhoVetoresInvertidos[cont], tamanhoVetoresAleatorios[cont] = len(vetoresOrdenados[cont]), len(vetoresInvertidos[cont]), len(vetoresAleatorios[cont])
    cont+=1

  listaX = [tamanhoVetoresOrdenados, tamanhoVetoresInvertidos, tamanhoVetoresAleatorios]
  return listaX

def retornarY(vetoresOrdenados, vetoresInvertidos, vetoresAleatorios, tipoOrganizacao):
  tempoOrdenados, tempoInvertidos, tempoAleatorios = list(range(len(vetoresOrdenados))), list(range(len(vetoresInvertidos))), list(range(len(vetoresAleatorios)))
  
  tamanho = len(vetoresOrdenados)
  if tamanho > len(vetoresInvertidos):
    tamanho = len(vetoresInvertidos)
  if tamanho > len(vetoresAleatorios):
    tamanho = len(vetoresAleatorios)

  cont = 0
  while cont < tamanho:
    #TIPO DE ALGORITMO DE ORDENAÇÃO
    if tipoOrganizacao == 'insertion':
      tempoOrdenados[cont], tempoInvertidos[cont], tempoAleatorios[cont] = retornarTempoExecucao(vetoresOrdenados[cont], 'insertion'), retornarTempoExecucao(vetoresInvertidos[cont], 'insertion'), retornarTempoExecucao(vetoresAleatorios[cont], 'insertion')
    elif tipoOrganizacao == 'selection':
      tempoOrdenados[cont], tempoInvertidos[cont], tempoAleatorios[cont] = retornarTempoExecucao(vetoresOrdenados[cont], 'selection'), retornarTempoExecucao(vetoresInvertidos[cont], 'selection'), retornarTempoExecucao(vetoresAleatorios[cont], 'selection')
    elif tipoOrganizacao == 'bubble':
      tempoOrdenados[cont], tempoInvertidos[cont], tempoAleatorios[cont] = retornarTempoExecucao(vetoresOrdenados[cont], 'bubble'), retornarTempoExecucao(vetoresInvertidos[cont], 'bubble'), retornarTempoExecucao(vetoresAleatorios[cont], 'bubble')
    cont+=1

  listaY = [tempoOrdenados, tempoInvertidos, tempoAleatorios]
  return listaY

def plotarGrafico(listaX, listaY, listaCor, titulo, labelX, labelY, listaLegenda):
  if len(listaX) == len(listaY):
    cont = 0
    while cont < len(listaX):
      plt.plot(listaX[cont], listaY[cont], linestyle='--', marker='o', color = listaCor[cont], markersize = 4)
      cont+=1

    plt.title(titulo, fontsize=15)
    plt.ylabel(labelY, fontsize=12)
    plt.xlabel(labelX, fontsize=12)
    plt.legend(listaLegenda, fontsize=10)
    axes = plt.gca()
    axes.yaxis.grid(b=True, color='black', alpha=0.3, linestyle='-.', linewidth=1)
    plt.figure(figsize=(10.5, 9))
    plt.show()

#VETORES ORDENADOS
vO = [list(range(100)), list(range(1000)), list(range(10000)), list(range(100000))]
#VETORES INVERTIDOS
vI = [list(reversed(list(range(100)))), list(reversed(list(range(1000)))), list(reversed(list(range(10000)))), list(reversed(list(range(100000))))]
#VETORES ALEATORIOS
vA = [sample(range(1, 101), 100), sample(range(1, 1001), 1000), sample(range(1, 10001), 10000), sample(range(1, 100001), 100000)]

valoresX, valoresYBubble, valoresYSelection, valoresYInsertion, cores = retornarX(vO, vI, vA), retornarY(vO, vI, vA, 'bubble'), retornarY(vO, vI, vA, 'selection'), retornarY(vO, vI, vA, 'insertion'), ['blue', 'red', 'black']

plotarGrafico(valoresX, valoresYBubble, cores, 'Gráfico Para Método Bubble', 'Tamanho', 'Tempo', ['Vetores Ordenados', 'Vetores Invertidos', 'Vetores Aleatorios'])
plotarGrafico(valoresX, valoresYSelection, cores, 'Gráfico Para Método Selection', 'Tamanho', 'Tempo', ['Vetores Ordenados', 'Vetores Invertidos', 'Vetores Aleatorios'])
plotarGrafico(valoresX, valoresYInsertion, cores, 'Gráfico Para Método Insertion', 'Tamanho', 'Tempo', ['Vetores Ordenados', 'Vetores Invertidos', 'Vetores Aleatorios'])
```
